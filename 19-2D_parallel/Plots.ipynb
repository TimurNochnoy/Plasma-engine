```python
import matplotlib.pyplot as plt
```

```python
def plot_for_S(filename):
    P = 16
    data = []
    
    with open(filename, "r") as f:
        for line in f:
            columns = line.split(' ')
            data.append([float(columns[0]), float(columns[1])])

    x = []
    y = []
        
    for i in range(P):
        for j in range(P):
            if data[j][0] == i + 1:
                x.append(data[j][0])
                y.append(data[j][1])
    
    for i in range(1, P):
        y[i] = y[0] / y[i]
    
    y[0] = 1.0
    
    x0 = []
    y1 = []
    y2 = []
    
    for i in range(P):
        x0.append(x[i])
        y1.append(y[i])
        y2.append(y[i] / x[i])
    
    plt.title('Анализ ускорения')
    plt.xlabel('Количество процессоров')
    plt.ylabel('Ускорение')
    plt.plot(x0, y1, label='200x100')
    plt.plot(x0, x0, label='идеальный случай')
    plt.plot(x0, y1, 'D')
    plt.xticks(ticks=x0, labels=[i+1 for i in range(P)])
    plt.yticks(ticks=x0, labels=[i+1 for i in range(P)]) 
    plt.grid()
    plt.legend()
    plt.savefig('speed up.png')
    plt.show()
    
    plt.title('Анализ эффективности')
    plt.xlabel('Количество процессоров')
    plt.ylabel('Эффективность')
    plt.plot(x0, y2, label='200x100')
    plt.plot(x0, y2, 'D')
    plt.xticks(ticks=x0, labels=[i+1 for i in range(P)])
    plt.grid()
    plt.legend()
    plt.savefig('efficiency.png')
    plt.show()
```

```python
plot_for_S("SUPERTIME.txt")
```

```python
plot_for_S("TIME.txt")
```

```python
plot_for_S("SMALL200X100.txt")
```

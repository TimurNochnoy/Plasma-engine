```python
import matplotlib.pyplot as plt
```

```python
def make_plot(filename):
    N = 0
    data = []
    
    y_value = 0.5
    
    with open(filename, "r") as f:
        for line in f:
            columns = line.split(' ')
            data.append([float(columns[0]), float(columns[1]), float(columns[2]), 
                         float(columns[3]), float(columns[4]), float(columns[5]), 
                         float(columns[6]), float(columns[7]), float(columns[8]),
                         float(columns[9]), float(columns[10]), float(columns[11])])
            N += 1

    z = []
    y = []
    rho = []
    v_r = []
    v_phi = []
    v_z = []
    H_phi = []
    H_z = []
    H_r = []
    e = []
    p = []
    P = []

    for i in range(N):
        if (data[i][1] == y_value):
            z.append(data[i][0])
            y.append(data[i][1])
            rho.append(data[i][2])
            v_r.append(data[i][3])
            v_phi.append(data[i][4])
            v_z.append(data[i][5])
            H_phi.append(data[i][6])
            H_z.append(data[i][7])
            H_r.append(data[i][8])
            e.append(data[i][9])
            p.append(data[i][10])
            P.append(data[i][11])
    
    plt.title('rho')
    plt.xlabel('z')
    plt.ylabel('rho')
    plt.plot(z, rho, label='y = ' + str(y_value))
    plt.grid()
    plt.legend()
    plt.show()
    
    plt.title('v_r')
    plt.xlabel('z')
    plt.ylabel('v_r')
    plt.plot(z, v_r, label='y = ' + str(y_value))
    plt.grid()
    plt.legend()
    plt.show()
    
    plt.title('v_phi')
    plt.xlabel('z')
    plt.ylabel('v_phi')
    plt.plot(z, v_phi, label='y = ' + str(y_value))
    plt.grid()
    plt.legend()
    plt.show()
    
    plt.title('v_z')
    plt.xlabel('z')
    plt.ylabel('v_z')
    plt.plot(z, v_z, label='y = ' + str(y_value))
    plt.grid()
    plt.legend()
    plt.show()
    
    plt.title('H_phi')
    plt.xlabel('z')
    plt.ylabel('H_phi')
    plt.plot(z, H_phi, label='y = ' + str(y_value))
    plt.grid()
    plt.legend()
    plt.show()
    
    plt.title('H_z')
    plt.xlabel('z')
    plt.ylabel('H_z')
    plt.plot(z, H_z, label='y = ' + str(y_value))
    plt.grid()
    plt.legend()
    plt.show()
    
    plt.title('H_r')
    plt.xlabel('z')
    plt.ylabel('H_r')
    plt.plot(z, H_r, label='y = ' + str(y_value))
    plt.grid()
    plt.legend()
    plt.show()
    
    plt.title('e')
    plt.xlabel('z')
    plt.ylabel('e')
    plt.plot(z, e, label='y = ' + str(y_value))
    plt.grid()
    plt.legend()
    plt.show()
    
    plt.title('p')
    plt.xlabel('z')
    plt.ylabel('p')
    plt.plot(z, p, label='y = ' + str(y_value))
    plt.grid()
    plt.legend()
    plt.show()
    
    plt.title('P')
    plt.xlabel('z')
    plt.ylabel('P')
    plt.plot(z, P, label='y = ' + str(y_value))
    plt.grid()
    plt.legend()
    plt.show()
```

```python
make_plot("data_MGD_CPP.txt")
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def make_plot(filename):
    N = 0
    data = []
    
    y_value = 0.5
    
    with open(filename, "r") as f:
        for line in f:
            columns = line.split(' ')
            data.append([float(columns[0]), float(columns[1]), float(columns[2]), 
                         float(columns[3]), float(columns[4]), float(columns[5]), 
                         float(columns[6]), float(columns[7]), float(columns[8]),
                         float(columns[9]), float(columns[10]), float(columns[11])])
            N += 1

    z = []
    y = []
    rho = []
    v_r = []
    v_phi = []
    v_z = []
    H_phi = []
    H_z = []
    H_r = []
    e = []
    p = []
    P = []

    for i in range(N):
        if (data[i][1] == y_value):
            z.append(data[i][0])
            y.append(data[i][1])
            rho.append(data[i][2])
            v_r.append(data[i][3])
            v_phi.append(data[i][4])
            v_z.append(data[i][5])
            H_phi.append(data[i][6])
            H_z.append(data[i][7])
            H_r.append(data[i][8])
            e.append(data[i][9])
            p.append(data[i][10])
            P.append(data[i][11])
    
    plt.title('rho')
    plt.xlabel('z')
    plt.ylabel('rho')
    plt.plot(z, rho, label='y = ' + str(y_value))
    plt.grid()
    plt.legend()
    plt.show()
    
    plt.title('v_r')
    plt.xlabel('z')
    plt.ylabel('v_r')
    plt.plot(z, v_r, label='y = ' + str(y_value))
    plt.grid()
    plt.legend()
    plt.show()
    
    plt.title('v_phi')
    plt.xlabel('z')
    plt.ylabel('v_phi')
    plt.plot(z, v_phi, label='y = ' + str(y_value))
    plt.grid()
    plt.legend()
    plt.show()
    
    plt.title('v_z')
    plt.xlabel('z')
    plt.ylabel('v_z')
    plt.plot(z, v_z, label='y = ' + str(y_value))
    plt.grid()
    plt.legend()
    plt.show()
    
    plt.title('H_phi')
    plt.xlabel('z')
    plt.ylabel('H_phi')
    plt.plot(z, H_phi, label='y = ' + str(y_value))
    plt.grid()
    plt.legend()
    plt.show()
    
    plt.title('H_z')
    plt.xlabel('z')
    plt.ylabel('H_z')
    plt.plot(z, H_z, label='y = ' + str(y_value))
    plt.grid()
    plt.legend()
    plt.show()
    
    plt.title('H_r')
    plt.xlabel('z')
    plt.ylabel('H_r')
    plt.plot(z, H_r, label='y = ' + str(y_value))
    plt.grid()
    plt.legend()
    plt.show()
    
    plt.title('e')
    plt.xlabel('z')
    plt.ylabel('e')
    plt.plot(z, e, label='y = ' + str(y_value))
    plt.grid()
    plt.legend()
    plt.show()
    
    plt.title('p')
    plt.xlabel('z')
    plt.ylabel('p')
    plt.plot(z, p, label='y = ' + str(y_value))
    plt.grid()
    plt.legend()
    plt.show()
    
    plt.title('P')
    plt.xlabel('z')
    plt.ylabel('P')
    plt.plot(z, P, label='y = ' + str(y_value))
    plt.grid()
    plt.legend()
    plt.show()
```

```python
make_plot("data_MGD_CPP.txt")
```

```python
def make_plot_2D(filename):
    N = 0
    data = []
    
    y_value = 0.5
    
    with open(filename, "r") as f:
        for line in f:
            columns = line.split(' ')
            data.append([float(columns[0]), float(columns[1]), float(columns[2]), 
                         float(columns[3]), float(columns[4]), float(columns[5]), 
                         float(columns[6]), float(columns[7]), float(columns[8]),
                         float(columns[9]), float(columns[10]), float(columns[11])])
            N += 1

    z = []
    y = []
    N_z = 0
    N_y = 0

    for i in range(N):
        if (data[i][0] not in z):
            z.append(data[i][0])
            N_z += 1
            
        if (data[i][1] not in y):
            y.append(data[i][1])
            N_y += 1
    
    
    rho = []
    v_r = []
    v_phi = []
    v_z = []
    H_phi = []
    H_z = []
    H_r = []
    e = []
    p = []
    P = []
    
    # N_z ---, N_y |||
    
    for i in range(N_z):
        rho_z = []
        v_r_z = []
        v_phi_z = []
        v_z_z = []
        H_phi_z = []
        H_z_z = []
        H_r_z = []
        e_z = []
        p_z = []
        P_z = []
        
        for j in range(i * N_y, i * N_y + N_y):
            rho_z.append(data[j][2])
            v_r_z.append(data[j][3])
            v_phi_z.append(data[j][4])
            v_z_z.append(data[j][5])
            H_phi_z.append(data[j][6])
            H_z_z.append(data[j][7])
            H_r_z.append(data[j][8])
            e_z.append(data[j][9])
            p_z.append(data[j][10])
            P_z.append(data[j][11])
        
        rho.append(rho_z)
        v_r.append(v_r_z)
        v_phi.append(v_phi_z)
        v_z.append(v_z_z)
        H_phi.append(H_phi_z)
        H_z.append(H_z_z)
        H_r.append(H_r_z)
        e.append(e_z)
        p.append(p_z)
        P.append(P_z)

    
    def func_Z(z, y):
        return rho[z][y]
    
    #plt.style.use('_mpl-gallery-nogrid')
    #plt.style.use('default')

    # make data with uneven sampling in x
    X, Y = np.meshgrid(z, y)
    #Z = (np.array(rho).T)[::-1]
    #print(Z)
    
    Z_value_lst = []
    Z_value_lst.append([(np.array(rho).T)[::-1],
                        (np.array(H_phi).T)[::-1]])
    Z_value_lst.append([(np.array(v_z).T)[::-1],
                        (np.array(v_r).T)[::-1]])
    
    fig, ax = plt.subplots(2, 2, figsize=(16, 13))
    title_name = [[r'  ($\rho$)', r'   ($H_{phi}$)'],
                  [ r' ($v_z$)', r'  ($v_{r}$)']]
    
    for i in range(0, 2):
        for j in range (0, 2):
            ax[i][j].pcolormesh(X, Y, Z_value_lst[i][j],
                                vmin=min([min(line) for line in Z_value_lst[i][j]]),
                                vmax=max([max(line) for line in Z_value_lst[i][j]]))
            ax[i][j].set_ylabel('y', fontsize = 14)
            ax[i][j].set_xlabel('z', fontsize = 14)
            ax[i][j].set_title(title_name[i][j], fontsize = 16)
            #ax[i][j].legend()
    
    plt.suptitle(r'$H_zS = 0.$', fontsize=20)
    fig.tight_layout()
    #plt.savefig('01.png')
    plt.show()
```

```python
make_plot_2D("data_MGD_CPP.txt")
```
